#### Data Cleaning

```python
# Ensuring datetime type
df["Submission Date"] = pd.to_datetime(df["Submission Date"])

# Splitting into separate columns
df["Date of Submission"] = df["Submission Date"].dt.date.astype(str)   # YYYY-MM-DD
df["Time of Submission"] = df["Submission Date"].dt.time.astype(str)   # HH:MM:SS

# Dropping original column
df = df.drop(columns=["Submission Date"])

# Moving new columns to the front
cols = ["Date of Submission", "Time of Submission"] + [c for c in df.columns if c not in ["Date of Submission", "Time of Submission"]]
df = df[cols]
```

```python
# Dropping unnecessary columns
df = df.drop(columns=["ID", "Your First Name", "Last Name", "Phone Number",
                     "Desired Date & Time of Event", "Address of Event", "City", "State", "Zip Code", 
                      "Total Number of Children, Adults or both Participating in the Event",
                     "Fire Safety/Fire Prevention Programs", "Additional Request or Questions?", 
                     "Available Sations for Tour", "Fire Safety/Fire Prevention Programs.1", "Available Stations for Tour", 
                     "End Time of the Event", "First Name", "IP Address", "Referrer", 
                      "Street Address of Event", "Organization Name", "Agency Name", "Type of Agency"])
display(df.sample(5))
print(df.columns.tolist())
```

```python
# Converting to datetime safely
df["Date of Submission"] = pd.to_datetime(df["Date of Submission"], errors="coerce")

# Filtering only 2025 and 2026 rows (BEFORE any other operations)
df = df[df["Date of Submission"].dt.year.isin([2025, 2026])].copy()

# Sorting by date in descending order
df = df.sort_values("Date of Submission", ascending=False)


word_map = {
    "zero": 0, "one": 1, "two": 2, "three": 3, "four": 4,
    "five": 5, "six": 6, "seven": 7, "eight": 8, "nine": 9,
    "ten": 10, "eleven": 11, "twelve": 12, "thirteen": 13,
    "fourteen": 14, "fifteen": 15, "sixteen": 16, "seventeen": 17,
    "eighteen": 18, "nineteen": 19, "twenty": 20,
    "thirty": 30, "forty": 40, "fifty": 50,
    "sixty": 60, "seventy": 70, "eighty": 80,
    "ninety": 90, "hundred": 100
}


def parse_word_number(text):
    t = text.strip().lower()
    return word_map.get(t, None)


def clean_participants(text):
    if pd.isna(text):
        return np.nan

    t = str(text).lower().replace(",", "").strip()

    nums = [float(x) for x in re.findall(r'\d+(?:\.\d+)?', t)]

    # Missing / invalid
    if not nums and not any(w in t for w in word_map):
        return np.nan

    # Homes excluded
    if "home" in t or "homes" in t:
        return np.nan

    # Word-only numbers
    word_num = parse_word_number(t)
    if word_num is not None:
        return float(word_num)

    # Over X
    over_pattern = re.search(r'\bover\s+(one|two|three|four|five|six|seven|eight|nine|ten|\d+)\s*(hundred)?', t)
    if over_pattern:
        base = over_pattern.group(1)
        mult = over_pattern.group(2)

        value = float(base) if base.isdigit() else float(word_map.get(base, np.nan))
        if mult:
            value *= 100
        return value

    # Explicit total
    total = re.search(r'(\d+)\s*total', t)
    if total:
        return float(total.group(1))

    # STRICT CLASS RULE
    strict_class_pattern = re.search(r'^\s*(\d+)\s+\d+\s*classes', t)
    if strict_class_pattern:
        return float(strict_class_pattern.group(1))

    # Room range per class
    room_class_pattern = re.search(
        r'(\d+)\s*(?:to|-)\s*(\d+)\s*(?:children|kids|students).*?(\d+)\s*classes',
        t
    )
    if room_class_pattern:
        low = float(room_class_pattern.group(1))
        high = float(room_class_pattern.group(2))
        classes = float(room_class_pattern.group(3))
        return ((low + high) / 2) * classes

    # Classroom inference
    classroom_pattern = re.search(r'\d+\s*(classrooms|classes).*?(?:so|about).*?(\d+)', t)
    if classroom_pattern:
        return float(classroom_pattern.group(2))

    # Group ranges
    group_pattern = re.search(r'(\d+)\s*(groups|classes|sessions).*?(\d+)\s*(?:-|to)\s*(\d+)?', t)
    if group_pattern:
        groups = float(group_pattern.group(1))
        if group_pattern.group(4):
            low = float(group_pattern.group(3))
            high = float(group_pattern.group(4))
            return groups * ((low + high) / 2)
        return groups * float(group_pattern.group(3))

    # Slash values
    slash_pattern = re.findall(r'(\d+)\s*/\s*(\d+)', t)
    if slash_pattern:
        values = [np.mean([float(a), float(b)]) for a, b in slash_pattern]
        return np.mean(values)

    # FIX 1: 16 (1st round) 12 (2nd round) → average rounds
    if ("round" in t or "1st" in t or "2nd" in t):
        round_values = re.findall(r'\b(\d+)\s*\(', t)
        if len(round_values) >= 2:
            return np.mean([float(x) for x in round_values])

    # FIX 2: 35 3–5 year olds + 15 1–2 year olds → SUM (ONLY when both age groups exist)
    age_group_values = re.findall(r'(\d+)\s*(?:3\s*[-–]\s*5|1\s*[-–]\s*2|3-5|1-2|3–5|1–2|year|yr|yo|old)', t)

    # Critical: only trigger when there are clearly multiple groups
    if len(re.findall(r'\d+\s*(?:3\s*[-–]\s*5|3-5|1\s*[-–]\s*2|1-2)', t)) >= 2:
        return sum(float(x) for x in age_group_values if x.isdigit())

    # Role-based sums
    role_keywords = [
        "children", "kids", "adults", "students",
        "teachers", "parents", "boys", "girls"
    ]
    if any(k in t for k in role_keywords) and len(nums) > 1:
        return sum(nums)

    # Numeric ranges
    if re.search(r'\d+\s*-\s*\d+', t):
        pairs = re.findall(r'(\d+)\s*-\s*(\d+)', t)
        return np.mean([np.mean([float(a), float(b)]) for a, b in pairs])

    return nums[0]


# Applying cleaning
df["Participants_Clean"] = df["Anticipated Number of Participants"].apply(clean_participants)

# Splitting into yearly dataframes
df_2025 = df[df["Date of Submission"].dt.year == 2025].copy()
df_2026 = df[df["Date of Submission"].dt.year == 2026].copy()

# Renaming columns
for d in [df_2025, df_2026]:
    d.rename(columns={
        "Date of Submission": "Submission Date",
        "Time of Submission": "Submission Time"
    }, inplace=True)
```
